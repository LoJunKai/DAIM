 - Select ~9k images from each category
 - Get rid of all images not shaped (?, ?, 3)

```python
import json

with open("coco_annos.json") as f:
    coco = json.load(f)

print(coco.keys())
print(coco['annotations'][0].keys())
print(coco['images'][0].keys())
```

```python
cat_list = ['tops', 'trousers', 'outerwear', 'dresses', 'skirts']

def count_cat(cat):
  num = 0
  for annot in coco['annotations']:
    cat_id = annot['category_id']
    if cat_id-1 == cat:
      num += 1

  return num
```

```python
for i in range(5):
  print("category " + cat_list[i], count_cat(i))

# category tops 18522
# category trousers 9586
# category outerwear 2153
# category dresses 9105
# category skirts 6522

# category tops 9258
# category trousers 9586
# category outerwear 2153
# category dresses 9105
# category skirts 6522
```

```python
import random
new_img = []
new_annot = []

for annot_in, annot_val in enumerate(coco['annotations']):
  img_id = annot_val['image_id']
  for i in coco['images']:
    if i['id'] == img_id:
      img = i
      break

  if annot_val['category_id'] == 1:
    # Tops
    if random.randint(0,1):
      continue
  
  new_annot.append(annot_val)
  new_img.append(img)
```

```python
coco['annotations'] = new_annot
coco['images'] = new_img
```

```python
with open('coco_annos_reduced.json', 'w') as f:
  json.dump(coco, f)
```

```python
''' Remove images with non-standard shapes 
    All images are (?, ?, 3)
'''

import numpy as np
from PIL import Image

with open("coco_annos.json") as f:
    coco = json.load(f)
```

```python
for image_in, image_dict in enumerate(coco['images']):
  pil_image = Image.open("image/" + image_dict['file_name'])
  image_arr = np.array(pil_image)
  try:
    if image_arr.shape[2] == 3:
      continue
  except IndexError:
    pass

  print(image_dict, image_arr.shape)
  coco['images'][image_in] = {}

  for i_in, i_val in enumerate(coco['annotations']):
    if i_val['image_id'] == image_dict['id']:
      coco['annotations'][i_in] = {}
      break
    print('not found')
```
