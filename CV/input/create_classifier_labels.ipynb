```
cat_list = ['tops', 'trousers', 'outerwear', 'dresses', 'skirts']
```

```
import json

with open('train.json') as f:
  file = json.load(f)
```

```
output = {}
for annos in file['annotations']:
  img_file_path = str(annos['image_id']) + '.jpg'
  cat_id = annos['category_id']

  if output.get(img_file_path, True) and cat_id != 1:
    output[img_file_path] = []

  if cat_id == 1:
    # tops
    continue

  output[img_file_path].append(cat_list[cat_id-1])
  # print(output[img_file_path])
```

```
for i in output.keys():
  if len(output[i]) > 1:
    print(output[i])
```

```
import csv
with open("train_labels.csv", "w") as f:
  writer = csv.writer(f)
  for key, value in output.items():
    writer.writerow([key, value])
```
